```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
class DiffusionWithConsumption:
    def __init__(self, n0_func, c_func):
        # Physical Constants
        self.D = 1.0     # Diffusion coefficient
        self.alpha = 1.0 # Consumption rate constant
        
        self.L = 1.0     # Length of the domain
        self.T = 1.0     # Time period
        
        # Concentrations
        self.n0_func = n0_func  # Nutrients Initial Distribution
        self.c_func = c_func    # Nutrient eaters
        
        
        # Space/Time Discretisation
        self.nx = 50
        self.nt = 100
        
        self.dx = self.L / (self.nx - 1)
        self.x = np.linspace(0, self.L, self.nx)
        self.t = np.linspace(0, self.T, self.nt)

        # Initial condition
        self.n0 = n0_func(self.x)

    def pde_rhs(self, t, n):
        'represents the right-hand side of the PDE using the finite difference method'
        dn_dt = np.zeros_like(n)
        
        # Apply the finite difference method
        for i in range(1, self.nx - 1):
            dn_dt[i] = self.D * (n[i + 1] - 2 * n[i] + n[i - 1]) / self.dx**2 - self.alpha * n[i] * self.c_func(self.x[i])
        
        # Boundary conditions
        dn_dt[0] = 0
        dn_dt[-1] = 0
        
        return dn_dt

    def solve(self):
        'Solve the PDE using Runge-Kutta method of order 5(4)'
        self.solution = solve_ivp(self.pde_rhs, [0, self.T], self.n0, t_eval=self.t, method='RK45')

    def plot(self, filename='diffusion_with_consumption.png', save= False):
        # Create the figure and subplots
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

        # Plot nutrient concentration over time
        for i in range(0, self.nt, self.nt // 10):
            # '//' operator is used for floor division
            ax1.plot(self.x, self.solution.y[:, i], label=f'n(x, t={self.solution.t[i]:.2f})')

        ax1.set_xlabel('x')
        ax1.set_ylabel('Nutrient Concentration n(x,t)')
        ax1.legend()
        ax1.set_title('Diffusion Equation with Spatially Varying Consumption')
        ax1.grid(True)

        # Plot the nutrient eater concentration
        ax2.plot(self.x, self.c_func(self.x), 'k--', label='c(x)')
        ax2.set_xlabel('x')
        ax2.set_ylabel('Nutrient Eater Concentration c(x)')
        ax2.legend()
        ax2.set_title('Nutrient Eater Concentration')
        ax2.grid(True)

        # Adjust layout
        plt.tight_layout()

        # Save the figure
        if save:
            plt.savefig(filename, dpi = 500)

        # Show plot
        plt.show()

# Initial condition function
def n0_func(x):
    return x / 1.0  # Normalized to length L
```

# Results for diffenrent Bacterial Concentrations

```python
# Nutrient eater concentration function
def c_func(x):
    c0 = 1.0
    return c0 * np.ones_like(x)

# Create an instance of the class and solve the PDE
diffusion_system = DiffusionWithConsumption(n0_func, c_func)
diffusion_system.solve()
diffusion_system.plot()
```

```python
# Nutrient eater concentration function
def c_func(x):
    L = 1.0  # Length of the domain, assumed here to normalize the function
    return np.exp(1 - x / L) / (L * (np.e - 1))

# Create an instance of the class and solve the PDE
diffusion_system = DiffusionWithConsumption(n0_func, c_func)
diffusion_system.solve()
diffusion_system.plot()
```
