```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

class DiffusionWithConsumption:
    def __init__(self, L, T, tau, n0_func, c_func, bc_type='dirichlet'):
        # Physical Constants
        self.L = L        # Length of the domain
        self.T = T        # Consumption time scale & GENERAL time period
        self.tau = tau    # Diffusion time scale
        
        self.D = L**2 / (2 * self.tau) # Diffusion coefficient
        self.alpha = 1.0 / T           # Rate of consumption

        # Concentrations
        self.n0_func = n0_func  # Nutrients Initial Distribution
        self.c_func = c_func    # Nutrient eaters
        
        # Space/Time Discretisation
        self.nx = 50
        self.nt = 100
        
        self.dx = self.L / (self.nx - 1)
        self.x = np.linspace(0, self.L, self.nx)
        self.t = np.linspace(0, self.T, self.nt)

        # Initial condition
        self.n0 = n0_func(self.x)
        
        # Boundary condition type
        self.bc_type = bc_type
        
        # Equation String
        self.eq_str = 'n/t = D n/x   n(x) c(n)'

    def pde_rhs(self, t, n):
        'represents the right-hand side of the PDE using the finite difference method'
        dn_dt = np.zeros_like(n)
        
        # Apply the finite difference method
        for i in range(1, self.nx - 1):
            d2n_dx2_i = (n[i + 1] - 2 * n[i] + n[i - 1]) / self.dx**2
            dn_dt[i] = self.D * d2n_dx2_i - self.alpha * n[i] * self.c_func(self.x[i])
        
        # Boundary conditions
        if self.bc_type == 'dirichlet':
            n[0] = 0
            n[-1] = 1
        elif self.bc_type == 'neumann':
            dn_dt[0] = 0
            dn_dt[-1] = 0
        
        return dn_dt

    def solve(self):
        'Solve the PDE using Runge-Kutta method of order 5(4)'
        self.solution = solve_ivp(self.pde_rhs, [0, self.T], self.n0, t_eval=self.t, method='RK45')

    def plot(self, filename=None):
        # Create the figure and subplots
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

        # Plot nutrient concentration over time
        for i in range(0, self.nt, self.nt // 10):
            # '//' operator is used for floor division
            ax1.plot(self.x, self.solution.y[:, i], label=f'n(x, t={self.solution.t[i]:.2f})')

        ax1.set_xlabel('x')
        ax1.set_ylabel('Nutrient Concentration n(x,t)')
        ax1.legend()
        ax1.set_title('Diffusion Equation with Spatially Varying Consumption')
        ax1.grid(True)
        
        # Plot the nutrient eater concentration
        line1, = ax2.plot(self.x, self.c_func(self.x), 'k--', label='c(x)')
        ax2.set_xlabel('x')
        ax2.set_ylabel('Bacteria Concentration c(x)')
        ax2.set_title('Nutrient Eater Concentration & Nutrient Flux')
        ax2.grid(True)

        # Calculate the absolute flux at the last time point
        n_last = self.solution.y[:, -1]
        # Compute flux (Phi = -D dn/dx)
        abs_flux = self.D * np.gradient(n_last, self.dx)
        
        # Create a second y-axis for the absolute flux
        ax3 = ax2.twinx()
        line2, = ax3.plot(self.x, abs_flux, 'r--', label=r'$|\Phi(x)| = D \, \frac{d}{dx} n(x)$')
        ax3.set_ylabel('Absolute Flux $|\Phi(x)|$', color='red')
        ax3.tick_params(axis='y', colors='red')
    
        # Add the flux value text inside the image
        diatom_flux = f'$|\\Phi(x=0)|$ = {abs_flux[0]:.6f}'
        ax3.text(0.05, -0.15, diatom_flux, transform=ax3.transAxes, fontsize=12, verticalalignment='bottom')


        # Combine legends from both axes
        lines = [line1, line2]
        labels = [line.get_label() for line in lines]
        ax2.legend(lines, labels, loc='upper center')

        # Adjust layout
        plt.tight_layout()

        # Save the figure
        if filename != None:
            plt.savefig(filename, dpi=500)

        # Show plot
        plt.show()

# Example usage:
def n0_func(x):
    return x / 10

def c_func(x):
    return 1 + 0.5 * np.sin(np.pi * x / 10)

# Create an instance of the class
solver = DiffusionWithConsumption(L=10, T=10, tau=1, n0_func=n0_func, c_func=c_func, bc_type='dirichlet')

# Solve the PDE
solver.solve()

# Plot the results
solver.plot()
```
