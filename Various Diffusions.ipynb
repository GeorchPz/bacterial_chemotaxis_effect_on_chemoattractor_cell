# Results for different Diffusion Speeds

$$
\frac{T_{diffusion}}{T_{consumption}} = 
\begin{cases} 
    \gg 1,  \;  \text{diffusion domination}   \\ 
    \sim 1, \;  \text{comparable time scales} \\
    \ll 1,  \;  \text{consumption domiation}
\end{cases}
\implies
T_d =
\begin{cases} 
10 \\ 
1 \\ 
0.1 
\end{cases}
\; \& \; T_c = 1
\implies
D = 
\begin{cases} 
1/20 \\ 
1/2 \\ 
5 
\end{cases}
$$

```python
syst_params = {
'L' : 1,  'T' : 1,
'Tc': 1,  'Td': 1,
'Nb': 1,
'nx': 50,   'nt': 100
}
L = syst_params['L']

def n0_linear(x):
    return x / L
def c_const(x):
    return np.ones_like(x)
```

```python
syst_params['Td'] = 10
syst_params['T']  = 3

diffusion_system = DiffusionWithConsumption1D(syst_params, n0_linear, c_const)
diffusion_system.solve_pde()
diffusion_system.solve_ode()
diffusion_system.plot()
```
